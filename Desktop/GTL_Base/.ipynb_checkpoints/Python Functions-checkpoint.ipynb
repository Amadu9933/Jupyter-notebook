#### Arguments

****We can also pass information into a function using arguments****

****Arguments are like inputs of the function****

****An argument is a value that is passed to the function when it is "called"****

#### Parameter

****A parameter is a variable that is passed into a function while defining****

****A parameter is like a placeholder for the actual value (argument)****

#### Argument Vs Parameter

****Arguments are the actual paramaeters/values that will be given as inputs to the function****

****Whereas parameters are considered 'formal' as they are just used to indicate how many values will be given to the function****

## Types of Arguments

###  Positional Arguments

****Simplest way of passing parameters to a function****

****Positional arguments are also called required arguments****

****In function definition each parameter is specified explicitly and is separated by commas****

#### Syntax

```python
def hello_arg(name):           # name is the parameter of the function
    print(name + ',' + ' how is the solar taxi ride')
```

### return Statement

****return statement is used to exit the function from where it was called****

```python
def multiplier():
    x = int(input('Enter 1st number: '))
    y = int(input('Enter 2nd number: '))
    return x*y
```

```python
multiplier()
```

```python
def adder():
    x = int(input('Enter 1st number: '))
    y = int(input('Enter 2nd number: '))
    return x+y
```

```python
adder()
```

```python
def another_multiplier(x,y):
    return x*y
```

```python
another_multiplier(5,15)
```

### Multiple Parameters

****We can set a number of parameters for function to accept (like in the above example we set the number of parameters = 2)****

****Every function expects the exact number of arguments as defined****

****If we have set the number of parameters to 3, then our function will work only if we provide it with correct number of arguments****

#### Syntax

****In the example below, we are passing 3 parametres while defining our "addition" function****

****It'll only work when we provide exactly 3 arguments when it is called****

```python
def addition(x,y,z):
    return x + y + z
```

```python
addition(4,5,15)
```

```python
addition(2,2)
```

```python
addition(2,2,2,2)
```

### 5) Default Parameter Value

****Value assigned to a paremeter in a function definition is known as default value or optional value****

****If function is called without changing the paraameter then it will return its default value****

****default value can be changed while calling the function****

```python
def addition(a,b=0):
    print('The value of a: ', a)
    print('The value of b: ', b)
    return a+b
```

```python
c = addition(4)
print(c)
d = addition(4,5)
print(d)
```
