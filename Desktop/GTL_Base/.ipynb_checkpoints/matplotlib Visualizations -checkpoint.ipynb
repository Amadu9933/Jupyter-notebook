## Line Plots
- It is visual representation to show changes over time
- The category data is along the horizontal axis, and all the value data is along the vertical axis


```python
new=np.arange(30)

#new

plt.plot(new)
```

```python
x=np.arange(6)
y=x*2
plt.plot(x,y,marker='o',color='green',linestyle='--')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Our Second Graph")
plt.savefig('my_line_plot.png')
plt.show()
```

## Subplots
- ax=plt.subplot(n,m,i) or plt.subplot(nmi)
- places ax at position i in n x m grid

```python

#plot 1
plt.figure(figsize=(8,5))
plt.subplot(2,2,1)
plt.plot(gas["Year"],gas['Canada'],'og-',label='canada')
plt.legend()
plt.title('gas prices of Canada')


#plot2

plt.subplot(2,2,2)
plt.plot(gas['Year'],gas['Italy'],'oy-',label='Italy')
plt.legend()
plt.title('gas price of Italy')

#plot3
plt.subplot(2,2,3)
plt.plot(gas['Year'],gas['France'],label='France')
plt.legend()
plt.title('gas price for France')
#plot4
plt.subplot(2,2,4)
plt.plot(gas['Year'],gas['Japan'],'*b-',label='Japan')
plt.legend()
plt.title('gas price for Japan')

#plt.tight_layout()
plt.show()
```

## Pie Chart
- Pie chart represents a part-to-whole relationship
- Each pie is composed of three components; area,central angle and arc lenght.


```python
plt.figure(figsize=(7,7))
marks=[100,129,54]
label=['monica','nesreen','theo']
cr=['yellow','#3A8D8F','red']
explode=[0,0,.1]
plt.pie(marks,labels=label,colors=cr,autopct='% .0f %%',
        explode=explode,textprops={'fontsize':18},shadow=True)
plt.title('Pupils Marks',fontsize=15)
plt.legend()
plt.savefig('pretech_pie.png')
plt.show()
```

```python
#Define the figure size
plt.figure(figsize=(7,7))
marks=[25,30,45,10]

#labels of the pie
labels=['Theo','Enoch','Emma','Rebirth']
plt.pie(marks,labels=labels)
plt.legend()
plt.show()


```

## Bar Chart
- A chart that uses bands of different heights to show different amounts so they can be distinguished
- A bar chart is used to represent comparative data

```python
x=[1,2,3,4,5]
y=[2,4,11,8,10]


plt.figure(figsize=(10,10))
plt.bar(x,y)
plt.bar(x,y, color='r')
plt.title('hello')
plt.show()
```

## Scatter Plots

Scatter plots are similar to regular plots but they show individuals points instead of values connected in series.

```python
m=np.linspace(0,20,40)
m
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y, marker='*',color='g')
plt.title('Our sine graph')
plt.xlabel('x axis')
plt.ylabel('sinX')
plt.grid()
plt.show()
```

```python
x=[10,30,4,25,7,8,9,3]
y=[12,15,18,12,3,32,11,8]
y=np.sin(x)
plt.scatter(x,y, marker='*',color='g')
plt.title('Our  graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

## Histogram
Histogram is a bar representation of data that varies over a range. It plots the height of the data belonging to a range on the y-axis and plots the range along the x-axis. It use a bar representation to show the data belonging to each range.
It divides the ranges into bins.

```python
from random import randint
x=[randint(15,80) for i in range(100)]
y=[20,30,40,50,60,70,80]

plt.hist(x, bins=y,color='g',ec='red')
plt.title('hist of ages')
plt.xlabel('age group')
plt.ylabel('no of students')
plt.show()
```

## Grouped Bar Chart

```python
plt.figure(figsize=(5,5))
labels=['Mon','Tue','Wed','Thu','Fri']
mobile_sales=[20,34,30,35,27]
pc_sales=[25,32,34,20,25]
x=np.arange(len(labels)) #the label location
width=0.35 #the width of the bars
a=plt.bar(x-width/2,mobile_sales,width,color='m', label='Mobile')
b=plt.bar(x+width/2,pc_sales,width,color='b', label='PC')
plt.ylabel('Total sales')
plt.title('Total sales by mobile and pc')
plt.xticks(x,labels)
plt.legend()
plt.bar_label(a)
plt.bar_label(b)
plt.savefig('groupedbar.png')
```

## Stacked Bar Chart

Some times if you want to see how a larger category is divided into smaller and what the relationship
<br> of each part has on the total amount 

```python
#data for 2016 olympics
plt.figure(figsize=(10,7))
x=['USA','GBR','China','Russia','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,17,26,19,15]
bottom_b=[a+b for a,b in zip(gold,silver)]
a=plt.bar(x,gold,label='gold',color='r')
b=plt.bar(x,silver,bottom=gold,label='silver',color='y')
c=plt.bar(x,bronze,bottom=bottom_b,label='bronze',color='g')
plt.bar_label(a,label_type='center')
plt.bar_label(b,label_type='center')
plt.bar_label(c,label_type='center')
plt.legend()
plt.title('olympic medals by Countries')
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.savefig('stackedbar.png')
plt.show()
```
