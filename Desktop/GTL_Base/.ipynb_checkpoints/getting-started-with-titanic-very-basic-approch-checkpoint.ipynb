<a id='sectiontwo-four'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center> Gender based survival rate</center></h2>

```python
import matplotlib.pyplot as plt
import seaborn

women = train_data[train_data['Sex']=='female']['Survived'].tolist()
rate_women = sum(women)/len(women)

men = train_data[train_data['Sex']=='male']['Survived'].tolist()
rate_men = sum(men)/len(men)
  
# declaring data
data = [rate_women, rate_men]
keys = ['Women', 'Men']
  
# declaring exploding pie
explode = [0.1, 0]
# define Seaborn color palette to use
palette_color = seaborn.color_palette('bright')
  
# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')
plt.title("Survival Ratio of Women vs Men")
# displaying chart
plt.show()
```

<a id='sectiontwo-nine'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center> Distribution of Suvival in Age and Fare </center></h2>

```python
cont_features = ['Age', 'Fare']
surv = train_data['Survived'] == 1

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(20, 20))
plt.subplots_adjust(right=1.5)

for i, feature in enumerate(cont_features):    
    # Distribution of survival in feature
    sns.distplot(train_data[~surv][feature], label='Not Survived', hist=True, color='#e74c3c', ax=axs[0][i])
    sns.distplot(train_data[surv][feature], label='Survived', hist=True, color='#2ecc71', ax=axs[0][i])
    
    # Distribution of feature in dataset
    sns.distplot(train_data[feature], label='Training Set', hist=False, color='#e74c3c', ax=axs[1][i])
    
    axs[0][i].set_xlabel('')
    axs[1][i].set_xlabel('')
    
    for j in range(2):        
        axs[i][j].tick_params(axis='x', labelsize=20)
        axs[i][j].tick_params(axis='y', labelsize=20)
    
    axs[0][i].legend(loc='upper right', prop={'size': 20})
    axs[1][i].legend(loc='upper right', prop={'size': 20})
    axs[0][i].set_title('Distribution of Survival in {}'.format(feature), size=20, y=1.05)

axs[1][0].set_title('Distribution of {} Feature'.format('Age'), size=20, y=1.05)
axs[1][1].set_title('Distribution of {} Feature'.format('Fare'), size=20, y=1.05)
        
plt.show()
```

# <p style="padding:10px;background-color:green;margin:10;color:white;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 1px 1px;overflow:hidden;font-weight:50">Preprocessing</p>

<a id='sectionthree-one'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center> Label Encoding</center></h2>

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

le_sex, le_emb = LabelEncoder(), LabelEncoder()
le_sex.fit(list(set(train_data['Sex'].tolist())))
le_emb.fit(list(set(train_data['Embarked'].tolist())))
train_data_copy['Sex'] = le_sex.transform(train_data_copy['Sex'].astype(str))
train_data_copy['Embarked'] = le_emb.transform(train_data_copy['Embarked'].astype(str))
test_data_copy['Sex'] = le_sex.transform(test_data_copy['Sex'].astype(str))
test_data_copy['Embarked'] = le_emb.transform(test_data_copy['Embarked'].astype(str))
```

<a id='sectionthree-two'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center> One-Hot Encoding</center></h2>

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(handle_unknown='ignore'), [5,6])], remainder='passthrough')
ct.fit(train_data_copy)
train_data_copy_ = np.array(ct.transform(train_data_copy))
test_data_copy_ = np.array(ct.transform(test_data_copy))
```

<a id='sectionthree-three'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center> Taking care of missing values</center></h2>

```python
train_data_copy['Age'] =  train_data_copy['Age'].fillna(train_data_copy['Age'].mean())
test_data_copy['Age'] =  test_data_copy['Age'].fillna(train_data_copy['Age'].mean())
```

<a id='sectionthree-four'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center>Normalization</center></h2>


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(train_data_copy_)
```

<a id='sectionthree-five'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center>Transforming into Feature X and Y</center></h2>

```python
X, y = sc.transform(train_data_copy_), train_data['Survived'].values.reshape(-1,1)
```

<a id = 'section-four'></a>
# <p style="padding:10px;background-color:green;margin:10;color:white;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 1px 1px;overflow:hidden;font-weight:50">Model Selection and Evaluation</p>

<a id='sectionfour-one'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center>Train test split</center></h2>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

<a id='sectionfour-two'></a>
<h2 style='background:green; border:10; border-radius: 10px; color:white'><center>Lazy model evaluation</center></h2>

```python
!pip install lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=0,ignore_warnings=False, custom_metric=None)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
```
